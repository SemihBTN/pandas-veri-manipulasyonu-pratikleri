```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def adjust_new(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        row["Performance_Score"]
```

```python
df["New_Performance_Score"] = df.apply(adjust_new , axis=1)
```

```python
df
```

```python
df["New_Name"] = df["Name"].apply(lambda x : x.replace("_"," "))
```

```python
def new_salary(row):
    if row["Performance_Score"] < 4:
        return row["Salary"] * 1.2
    elif row["Performance_Score"] >= 4:
        return row["Salary"] * 1.1 
    else:
        return row["Salary"]
```

```python
df["New_Salary"] = df.apply(new_salary , axis=1)
```

```python
df
```

```python
def tecrubee(row):
    if row["Experience"] <= 5:
        return "Junior"
    elif 5 < row["Experience"] <= 15:
        return "Mid-Senior"
    elif row["Experience"] > 15:
        return "Senior"
    else:
        return "Noob"
```

```python
df["Experience_Level"] = df.apply(tecrubee , axis=1)
```
